```python
# Grado en Ingeniería Informática - Universidad de La Laguna
# Normativa y Regulación
# Eduardo González Gutiérrez
# Alu0101461588

import gdown
import pandas as pd
import matplotlib.pyplot as plt

# Descargar los archivos desde Google Drive usando gdown
renta_url = 'https://drive.google.com/uc?id=1RGbGqlvft6YyREyvbTjRZ_Xf_5Emd0rU'
esperanza_url = 'https://drive.google.com/uc?id=1vMd7yZJY0GvT3vjUoap5BLkxQxeobn9-'

# Definir rutas de destino
renta_dest = 'renta_media.csv'
esperanza_dest = 'esperanza_vida.csv'

# Descargar los archivos
gdown.download(renta_url, renta_dest, quiet=False)
gdown.download(esperanza_url, esperanza_dest, quiet=False)

# Cargar los datasets con el tipo de datos específico para "Total"
renta_df = pd.read_csv(renta_dest, sep='\t', encoding='ISO-8859-1', dtype={'Total': str})
esperanza_df = pd.read_csv(esperanza_dest, sep='\t', encoding='ISO-8859-1')

# Limpiar los nombres de las columnas para evitar problemas con los acentos y espacios
renta_df.columns = renta_df.columns.str.strip()
esperanza_df.columns = esperanza_df.columns.str.strip()

# Función para mostrar las comunidades autónomas
def mostrar_comunidades(df):
    comunidades = df['Comunidades y Ciudades Autónomas'].unique()
    for i, comunidad in enumerate(comunidades):
        print(f"{i+1}. {comunidad}")
    while True:
        try:
            seleccion = int(input("\nIntroduce el número de la comunidad autónoma que deseas seleccionar: "))
            if 1 <= seleccion <= len(comunidades):
                print(f"Has seleccionado: {comunidades[seleccion-1]}\n")
                return comunidades[seleccion-1]
            else:
                print("Por favor, introduce un número válido.")
        except ValueError:
            print("Por favor, introduce un número válido.")

# Función para mostrar los años disponibles para una comunidad seleccionada
def mostrar_anos(df, comunidad_seleccionada):
    df_comunidad = df[df['Comunidades y Ciudades Autónomas'] == comunidad_seleccionada]
    anos = df_comunidad['Periodo'].unique()
    for i, ano in enumerate(anos):
        print(f"{i+1}. {ano}")
    while True:
        try:
            seleccion = int(input("\nIntroduce el número del año que deseas seleccionar: \n"))
            if 1 <= seleccion <= len(anos):
                print(f"Has seleccionado el año: {anos[seleccion-1]}\n")
                return anos[seleccion-1]
            else:
                print("Por favor, introduce un número válido.")
        except ValueError:
            print("Por favor, introduce un número válido.")

# Función para obtener la esperanza de vida para una comunidad y año seleccionados
def obtener_esperanza_vida(df, comunidad_seleccionada, ano_seleccionado):
    comunidad_sin_numero = comunidad_seleccionada.split(" ", 1)[-1]
    esperanza = df[(df['Comunidades y Ciudades Autónomas'].str.contains(comunidad_sin_numero)) & (df['Periodo'] == ano_seleccionado)]['Total'].values
    if len(esperanza) > 0:
        esperanza_vida_str = esperanza[0].replace(',', '.')
        return float(esperanza_vida_str)
    else:
        return None

# Función para obtener la renta media para una comunidad y año seleccionados
def obtener_renta_media(df, comunidad_seleccionada, ano_seleccionado):
    # Extraer el nombre de la comunidad sin el número inicial ya que el DataSet esta de la forma "01 Canarias", es decir, las Comunidades están enumeradas
    comunidad_sin_numero = comunidad_seleccionada.split(" ", 1)[-1]

    # Filtrar el DataFrame para la comunidad y año seleccionados
    renta = df[
        (df['Comunidades y Ciudades Autónomas'].str.contains(comunidad_sin_numero)) &
        (df['Periodo'] == ano_seleccionado)
    ]['Total'].values

    if len(renta) > 0:
        # Convertir el valor a string (ya asegurado en la lectura) y limpiar el formato quitando los puntos
        renta_str = renta[0].replace('.', '').replace(',', '.')

        # Convertir a número entero de la forma "10490"
        try:
            renta_valor = int(renta_str)
            return renta_valor
        except ValueError:
            print(f"Error al convertir la renta: {renta_str}")
            return None
    else:
        return None

# Función para crear la gráfica de la relación entre renta y esperanza de vida
def graficar_renta_vs_esperanza(renta_df, esperanza_df, comunidad_seleccionada):
    # Limpia y estandariza los nombres de las comunidades en ambos DataFrames
    renta_df['Comunidades y Ciudades Autónomas'] = renta_df['Comunidades y Ciudades Autónomas'].str.strip().str.lower()
    esperanza_df['Comunidades y Ciudades Autónomas'] = esperanza_df['Comunidades y Ciudades Autónomas'].str.strip()

    # Modifica el formato de las comunidades en el DataFrame de esperanza para compararlas correctamente
    esperanza_df['Comunidades y Ciudades Autónomas'] = esperanza_df['Comunidades y Ciudades Autónomas'].str.split(' ', n=1).str[1].str.lower()

    # Limpia el nombre de la comunidad seleccionada
    comunidad_seleccionada = comunidad_seleccionada.strip().lower()

    # Filtra los datos de la comunidad seleccionada en ambos DataFrames
    renta_comunidad = renta_df[renta_df['Comunidades y Ciudades Autónomas'] == comunidad_seleccionada]
    esperanza_comunidad = esperanza_df[esperanza_df['Comunidades y Ciudades Autónomas'] == comunidad_seleccionada]

    # Si no hay datos de esperanza de vida, muestra un mensaje y termina la función
    if esperanza_comunidad.empty:
        print(f"\nNo se encontraron datos de esperanza de vida para la comunidad '{comunidad_seleccionada}'.")
        return None

    # Limpia y estandariza los valores de la columna 'Periodo'
    renta_comunidad['Periodo'] = renta_comunidad['Periodo'].astype(str).str.strip()
    esperanza_comunidad['Periodo'] = esperanza_comunidad['Periodo'].astype(str).str.strip()

    # Une los dos DataFrames en base al periodo y selecciona las columnas necesarias
    df_combinado = pd.merge(
        renta_comunidad[['Periodo', 'Total']],
        esperanza_comunidad[['Periodo', 'Total']],
        on='Periodo',
        suffixes=('_renta', '_esperanza')
    )

    # Limpia y convierte los valores de renta a enteros (elimina separadores de miles y decimales)
    df_combinado['Total_renta'] = df_combinado['Total_renta'].astype(str).str.replace('.', '').str.replace(',', '').astype(int)
    # Limpia y convierte los valores de esperanza de vida a flotantes
    df_combinado['Total_esperanza'] = df_combinado['Total_esperanza'].astype(str).str.replace(',', '.').astype(float)

    # Devuelve el DataFrame combinado con los datos procesados
    return df_combinado

# Función para crear la gráfica de la relación entre renta y esperanza de vida uniendo los puntos con líneas
def graficar_relacion_renta_esperanza(df_combinado, comunidad_seleccionada):
    # Configura el tamaño de la figura
    plt.figure(figsize=(10, 6))

    # Crea un gráfico de dispersión para los datos
    plt.scatter(df_combinado['Total_renta'], df_combinado['Total_esperanza'], color='blue', alpha=0.7, label='Datos')

    # Dibuja una línea que conecta los puntos, mostrando la tendencia
    plt.plot(df_combinado['Total_renta'], df_combinado['Total_esperanza'], color='blue', linestyle='-', alpha=0.5, label='Tendencia')

    # Añade etiquetas de los años (Periodo) a cada punto del gráfico
    for i, row in df_combinado.iterrows():
        plt.text(row['Total_renta'], row['Total_esperanza'], str(row['Periodo']), fontsize=9, ha='right', va='bottom')

    # Personaliza la gráfica con título, etiquetas y cuadrícula
    plt.title(f'Relación entre Renta Media y Esperanza de Vida en {comunidad_seleccionada.capitalize()}', fontsize=14)
    plt.xlabel('Renta Media (€)', fontsize=12)
    plt.ylabel('Esperanza de Vida (años)', fontsize=12)
    plt.grid(alpha=0.4)
    plt.legend()  # Añade la leyenda
    plt.tight_layout()  # Ajusta el diseño de la gráfica
    plt.show()  # Muestra la gráfica

# Función principal del programa
def main():
    # Bucle principal que mantiene activo el programa
    while True:
        # Menú principal para que el usuario seleccione una acción
        print("\n")
        print("\n----------------------------------------------------------------------------------------------------------------------------")
        print("Bienvenido al programa de análisis de la relación entre la renta media y la esperanza de vida.")
        print("1. Elegir comunidad autónoma")
        print("2. Salir del programa")
        print("----------------------------------------------------------------------------------------------------------------------------\n")

        try:
            # Captura la opción seleccionada por el usuario en el menú principal
            opcion_menu = int(input("Introduce el número de la opción que deseas seleccionar: "))
            if opcion_menu == 1: # Si se elige la opción 1, se muestra la lista de comunidades disponibles
                comunidad_seleccionada = mostrar_comunidades(renta_df)

                # Submenú para realizar diferentes acciones con la comunidad seleccionada
                while True:
                    print("\n¿Qué deseas hacer?")
                    print("1. Ver renta y esperanza de vida para un año concreto.")  # Opción para mostrar datos de un año específico
                    print("2. Ver la gráfica de relación entre renta y esperanza de vida.")  # Opción para generar una gráfica
                    print("3. Volver al menú principal.")  # Opción para regresar al menú principal

                    try:
                        # Captura la opción seleccionada en el submenú
                        opcion = int(input("\nIntroduce el número de la opción que deseas seleccionar: "))

                        if opcion == 1:
                            # Muestra los años disponibles para la comunidad seleccionada
                            ano_seleccionado = mostrar_anos(renta_df, comunidad_seleccionada)

                            # Obtiene los valores de esperanza de vida y renta media para el año seleccionado
                            esperanza_vida = obtener_esperanza_vida(esperanza_df, comunidad_seleccionada, ano_seleccionado)
                            renta_media = obtener_renta_media(renta_df, comunidad_seleccionada, ano_seleccionado)

                            # Verifica si se encontraron los datos y los muestra, o informa si faltan
                            if esperanza_vida is not None and renta_media is not None:
                                print(f"\nLa esperanza de vida para la comunidad '{comunidad_seleccionada}' en el año {ano_seleccionado} es: {esperanza_vida:.2f} años")
                                print(f"La renta media para la comunidad '{comunidad_seleccionada}' en el año {ano_seleccionado} es: {renta_media} €")

                            else:
                                print(f"No se encontraron datos para la comunidad '{comunidad_seleccionada}' en el año {ano_seleccionado}.")

                        elif opcion == 2: # Combina y prepara los datos para generar la gráfica
                            df_combinado = graficar_renta_vs_esperanza(renta_df, esperanza_df, comunidad_seleccionada)

                            if df_combinado is not None: # Genera y muestra la gráfica
                                graficar_relacion_renta_esperanza(df_combinado, comunidad_seleccionada)

                        elif opcion == 3: # Rompe el bucle del submenú y regresa al menú principal
                            break

                        else:
                            print("Por favor, selecciona una opción válida.")
                    except ValueError:
                        print("Por favor, introduce un número válido.")

            elif opcion_menu == 2: # Finaliza el programa si el usuario elige salir
                print("¡Gracias por usar el programa! Saliendo...")
                break

            else:
                print("Por favor, selecciona una opción válida.")

        except ValueError:
            # Maneja errores si el usuario no introduce un número válido
            print("Por favor, introduce un número válido.")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```
